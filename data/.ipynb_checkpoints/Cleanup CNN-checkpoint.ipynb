```python
import nltk # language processing
import os # operation system
import re
```

```python
def createFile(newssite,categorie,title,content):
    path = "/home/retkowski/data/" + newssite + "/" + categorie + "/"
    if not os.path.exists(path):
        os.makedirs(path)

    filename = title + '.txt'
    if not (os.path.exists(filename)):
        with open(os.path.join(path, filename), 'wb') as temp_file:
            temp_file.write(str.encode(content))
            temp_file.close()
```

```python
def get_categories_from_folder(site):
    categories = []
    for root, dirs, files in os.walk('/home/retkowski/data/'+site):
        for name in dirs:
            if (name.find('.',0,1) == -1):
                categories.append(name)
    return categories
print(get_categories_from_folder('cnn'))
```

```python
site = 'cnn'
categories = get_categories_from_folder(site)

cleanup_data = []
    
for categorie in categories:      
    path = "/home/retkowski/data/"+site+"/"+categorie
    if os.path.exists(path):
        count = 0 
            
    for file in os.listdir(path):                     
        if file.endswith(".txt"):
            with open(path+"/"+file) as f:
                cleanup_data.append(f.read())
                count += 1
                f.close()
                
print(cleanup_data[3])
```

```python
text = cleanup_data[3]
text = re.sub(r"(\r\n?|\n)*Read More.*(\r\n?|\n)","",text)
text = re.sub(r"(\r\n?|\n)*.*JUST WATCHED.*(\r\n?|\n)","",text)
text = re.sub(r"(\r\n?|\n)*\(CNN\).*(\r\n?|\n)","",text)
print(text)
```
